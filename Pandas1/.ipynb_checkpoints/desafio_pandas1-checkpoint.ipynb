```python
import pandas as pd
```

```python
url = 'https://raw.githubusercontent.com/alura-cursos/pandas-conhecendo-a-biblioteca/main/desafios/alunos.csv'

dados = pd.read_csv(url)

dados.head()
```

```python
dados.isnull()
```

```python
dados = dados.fillna(0)

dados
```

```python
remocao = dados.query('Notas == 0').index
```

```python
dados.drop(remocao, axis=0, inplace=True)
```

```python
dados
```

```python
selecao2 = dados.query('Nome == "Alice" and Nome == "Carlos"').index

dados.drop(selecao2, axis=0, inplace=True)

dados
```

```python
aprovados = dados['Aprovado'] == True
```

```python
alunos_aprovados = dados[aprovados]
```

```python
alunos_aprovados

alunos_aprovados.replace(7.0, 8.0)
```

```python
alunos_aprovados.to_csv('alunos_aprovados.csv', index=False, sep=';')
```

```python
dados = dados.fillna(0)
```

```python
dados['Pontos_extra'] = 0.4*dados['Notas']

dados.head()
```

```python
dados['Nota_final'] = dados['Notas'] + dados['Pontos_extra']

dados.head()
```

```python
dados['Aprovado_final'] = dados['Nota_final'].apply(lambda x: True if x >=6 else False)

dados.head()
```

```python
selecao3 = dados.query('Aprovado_final == False').index

selecao3
```

```python
alunos_aprovados_2 = dados

dados.head()
```
