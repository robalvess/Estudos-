```python
url = 'https://raw.githubusercontent.com/alura-cursos/pandas-conhecendo-a-biblioteca/main/base-de-dados/aluguel.csv'

dados = pd.read_csv(url, sep=';')

dados = dados.fillna(0)
```

```python
#Criando coluna

dados['Valor_por_mes'] = dados['Valor'] + dados['Condominio']


dados.head()
```

```python
dados['Valor_por_ano'] = 12*dados['Valor_por_mes'] + dados['IPTU']

dados.head()
```

```python
dados['Descricao'] = dados['Tipo'] + ' em ' + dados['Bairro'] +  ' com ' +  dados['Quartos'].astype(str) + ' quartos ' +  ' e ' + dados['Vagas'].astype(str) + ' vaga(s) de garagem. '

dados.head()
```

```python
dados['Possui_suite'] = dados['Suites'].apply(lambda x : 'Sim' if x > 0 else 'Não')

dados.head()
```

```python
dados.to_csv('DADOS_DEV', index='False', sep=';')
```
